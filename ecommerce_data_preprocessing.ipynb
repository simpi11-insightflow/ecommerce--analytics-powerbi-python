```python
#E-commerce Data Analysis Project
#Import+Load Data
import pandas as pd
```

```python
path="C:/Users/simpi/Downloads/archive/"
orders=pd.read_csv(path+"olist_orders_dataset.csv")
orders.head()
```

```python
customers=pd.read_csv(path+"olist_customers_dataset.csv")
order_items=pd.read_csv(path+"olist_order_items_dataset.csv")
payments=pd.read_csv(path+"olist_order_payments_dataset.csv")
```

```python
customers.head()
order_items.head()
payments.head()
```

```python
orders=orders[orders['order_status']=='delivered']
```

```python
orders=orders.dropna(subset=['order_id','customer_id'])
```

```python
order_items=order_items.dropna(subset=['order_id','price'])
```

```python
payments=payments.dropna(subset=['order_id','payment_value'])
```

```python
orders['order_purchases_timestamp']=pd.to_datetime(orders['order_purchase_timestamp'])
```

```python
#Merging orders and customers
df=pd.merge(orders,customers,on='customer_id',how='left')
```

```python
df[['order_id','customer_id','customer_city']].head(10)
```

```python
#Add order items
df=pd.merge(df,order_items,on='order_id',how='left')
```

```python
df=pd.merge(df,payments,on='order_id',how='left')
```

```python
df.shape
df.head()
```

```python
#Grouping
payments_grouped =(
payments.groupby('order_id')
['payment_value'].sum().reset_index()
)
```

```python
payments_grouped.head("10")
```

```python
#Feature Engineering 
#force conversion
df['order_delivered_customer_date']=(
pd.to_datetime(df['order_delivered_customer_date'])
)

df['order_purchase_timestamp']=(
pd.to_datetime(df['order_purchase_timestamp'])
)
```

```python
#delivery time
df['delivery_time']=(
    
df['order_delivered_customer_date'] -
df['order_purchase_timestamp']
).dt.days
```

```python
#force conversion
df['order_delivered_customer_date']=(
pd.to_datetime(df['order_delivered_customer_date'],errors='coerce')
)
df['order_estimated_delivery_date']=(
pd.to_datetime(df['order_estimated_delivery_date'],errors='coerce')
)
```

```python
#Delay
df['delay']=(
   
df['order_delivered_customer_date']-
df['order_estimated_delivery_date']
).dt.days
```

```python
#order month
df['order_month']=(
df['order_purchase_timestamp'].dt.to_period('M')
)
```

```python
#total order value
df.rename(columns={'payment_value':'total_order_value'},inplace=True)
```

```python
#Business Analysis
#total revenue
df['total_order_value'].sum()
```

```python
#monthly revenue trend
df.groupby('order_month')['total_order_value'].sum()
```

```python
#Average delivery time
df['delivery_time'].mean()
```

```python
#late delivery %
(df['delay']>0).mean()*100
```

```python
#top cities by order
df['customer_city'].value_counts().head(10)
```

```python
#Payment type analysis
df['payment_type'].value_counts()
```
